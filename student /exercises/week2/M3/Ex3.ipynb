```python
import numpy as np
from scipy import stats

# Step 1: Phone usage data (in hours)
phone_usage = [2.5, 3.2, 4.0, 3.8, 3.0, 3.5, 4.2, 2.8, 3.6, 3.1, 3.9, 3.4, 4.1, 3.3, 3.7]

# Step 2: Null hypothesis mean (H0)
population_mean = 3  # claim: average phone usage is 3 hours/day

# Step 3: Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(phone_usage, population_mean)

# Step 4: One-tailed test (H1: mean > 3)
p_value_one_tailed = p_value / 2 if t_stat > 0 else 1 - (p_value / 2)

# Step 5: Print results
print("Sample mean phone usage:", round(np.mean(phone_usage), 2), "hours")
print("t-statistic:", round(t_stat, 3))
print("One-tailed p-value:", round(p_value_one_tailed, 4))

# Step 6: Conclusion
alpha = 0.05
if p_value_one_tailed < alpha:
    print("Reject H0: People use their phones more than 3 hours on average.")
else:
    print("Fail to reject H0: No evidence that people use phones more than 3 hours.")
```
