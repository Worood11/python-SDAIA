## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv('StudentPerformanceFactors.csv')
df.head()
```

## Data Cleaning

```python
# Missing values
df.isnull().sum()

# Remove duplicates
df = df.drop_duplicates()

```

## Exam Score Distribution

```python
sns.histplot(df['Exam_Score'], kde=True)
plt.title("Overall Exam Score Distribution")
plt.show()
```

# Attendance vs Exam Score

### Check Data Type

```python
df[['Attendance', 'Exam_Score']].dtypes
```

### Visualization

```python
plt.scatter(df['Attendance'], df['Exam_Score'])
plt.xlabel('Attendance')
plt.ylabel('Exam Score')
plt.title('Attendance vs Exam Score')
plt.show()
```

### Hypothesis Test (Correlation)

```python
r, p = stats.pearsonr(df['Attendance'], df['Exam_Score'])
print("Correlation:", r)
print("p-value:", p)
```

```python
if p < 0.05:
    print("Reject H0 → 'Attendance affects exam score")
else:
    print("Fail to reject H0 → No strong evidence")
```

# Gender vs Exam Score

### Check Data Type

```python
df['Gender'].value_counts()
```

### Visualization

```python
df.boxplot(column='Exam_Score', by='Gender')
plt.title('Gender vs Exam Score')
plt.suptitle('')
plt.show()
```

### Independent t-test

```python
male = df[df['Gender'] == 'Male']['Exam_Score']
female = df[df['Gender'] == 'Female']['Exam_Score']

t, p = stats.ttest_ind(male, female)
print("t-stat:", t)
print("p-value:", p)
```

```python
if p < 0.05:
    print("Reject H0 → There is a statistically significant difference in exam scores between genders")
else:
    print("Fail to reject H0 → No statistically significant evidence of a difference between genders")
```

# Parental Education vs Exam Score

### Check Data Type

```python
df['Parental_Education_Level'].value_counts()
```

### Visualization

```python
df.boxplot(column='Exam_Score', by='Parental_Education_Level')
plt.xticks(rotation=45)
plt.show()
```

### One-way ANOVA

```python
groups = [group['Exam_Score'].values for _, group in df.groupby('Parental_Education_Level')]

f, p = stats.f_oneway(*groups)
print("F-stat:", f)
print("p-value:", p)
```

```python
if p < 0.05:
    print("Reject H0 → 'Parental Education Level affects exam score")
else:
    print("Fail to reject H0 → No strong evidence")
```

# Internet Access vs Exam Score

### Check Data Type

```python
df['Internet_Access'].value_counts()
```

### Visualization

```python
df.boxplot(column='Exam_Score', by='Internet_Access')
plt.show()
```

### Independent t-test

```python
g1 = df[df['Internet_Access'] == 'Yes']['Exam_Score']
g2 = df[df['Internet_Access'] == 'No']['Exam_Score']

t, p = stats.ttest_ind(g1, g2)
print(t, p)
```

```python
if p < 0.05:
    print("Reject H0 → 'Internet Access affects exam score")
else:
    print("Fail to reject H0 → No strong evidence")
```

## Motivation_Level vs Exam Score

### Check Data Type

```python
df['Motivation_Level'].value_counts()
```

### Visualization

```python
df.boxplot(column='Exam_Score', by='Motivation_Level')
plt.title('Motivation Level vs Exam Score')
plt.suptitle('')
plt.xlabel('Motivation Level')
plt.ylabel('Exam Score')
plt.show()
```

### One-way ANOVA

```python
groups = [group['Exam_Score'].values for _, group in df.groupby('Motivation_Level')]

f_stat, p_value = stats.f_oneway(*groups)

print("F-stat:", f_stat)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject H0 → Motivation affects exam score")
else:
    print("Fail to reject H0 → No strong evidence")
```

# Final Conclusion

The analysis shows that academic performance is strongly influenced by behavioral and environmental factors. Hours studied and motivation level are the most significant predictors of exam scores. Attendance and parental education level also play important roles, indicating both personal effort and socio-economic background contribute to student success. In contrast, gender does not have a significant impact on performance.
