
# 🌍 Global Mental Health Crisis Index 2026

## Project Overview
This notebook performs **Exploratory Data Analysis (EDA)** on the Global Mental Health Crisis Index dataset.

### Objectives
- Understand global mental health patterns
- Analyze relationships between social, economic, and health variables
- Identify key drivers of mental health crisis severity

### Key Questions
1. How does social media usage relate to mental health?
2. Are wealthier countries better equipped for mental health?
3. What is the impact of COVID-19 on mental health?
4. How does healthcare capacity affect crisis levels?
5. Are there inequalities across income groups?


```python

# Standard libraries
from pathlib import Path

# Installed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import kagglehub
import os

# Configuration
pd.set_option('display.max_columns', None)
np.random.seed(42)

# Paths
path = kagglehub.dataset_download("alitaqishah/global-mental-health-crisis-index-2026")
```

## Data Loading & Validation

```python
DATA_PATH = os.path.join(path,'Global_Mental_Health_Crisis_Index_2026.csv')

df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
df.head()
```

```python

df.info()
```


## 📱 Use Case 1: Social Media vs Mental Health

We explore whether increased social media usage is associated with higher depression rates.


```python
plt.figure()
plt.scatter(df['social_media_hours_daily'], df['depression_pct'])
plt.xlabel("Social Media Hours per Day")
plt.ylabel("Depression Rate (%)")
plt.title("Social Media Usage vs Depression Rate")
plt.show()
```


## 💰 Use Case 2: GDP vs Mental Health Crisis

We examine whether wealthier countries have lower crisis scores.


```python
plt.figure()
plt.scatter(df['gdp_per_capita_usd'], df['mh_crisis_index'])
plt.xlabel("GDP per Capita")
plt.ylabel("Mental Health Crisis Index")
plt.title("GDP vs Crisis Index")
plt.show()
```


## 🦠 Use Case 3: COVID-19 Impact

Analyzing how pandemic-related increases affected crisis levels.


```python

plt.figure()
plt.scatter(df['covid_mh_increase_pct'], df['mh_crisis_index'])
plt.xlabel("COVID Impact (%)")
plt.ylabel("Crisis Index")
plt.title("COVID Impact vs Crisis Index")
plt.show()
```


## 🏥 Use Case 4: Healthcare Capacity

Do more psychiatrists reduce crisis severity?


```python

plt.figure()
plt.scatter(df['psychiatrists_per100k'], df['mh_crisis_index'])
plt.xlabel("Psychiatrists per 100k")
plt.ylabel("Crisis Index")
plt.title("Healthcare Capacity vs Crisis")
plt.show()
```


## 🌎 Use Case 5: Income Group Inequality

Comparing crisis index across income groups.


```python

df.groupby('income_group')['mh_crisis_index'].mean().plot(kind='bar')
plt.xlabel("Income Group")
plt.ylabel("Average Crisis Index")
plt.title("Crisis Index by Income Group")
plt.show()
```


## 📊 Summary & Key Insights

- Higher social media usage shows a positive relationship with depression
- Wealthier countries tend to have slightly lower crisis scores
- COVID-19 significantly increased mental health burden
- More psychiatrists generally correlate with lower crisis levels
- Lower-income countries face higher crisis severity
